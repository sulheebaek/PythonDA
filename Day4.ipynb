## 실습: 타이타닉 생존자 경향 분석
* titanic.csv 파일을 읽어들여 DataFrame 객체인 titanic을 생성하고 
* agerange 열을 추가, agerange 열은 age 열에 포함된 나이를 0~15, 16~60, 61~120세로 구분한 범주형 데이터가 되도록한다.
* 연령대와 승객등급별 생존자의 합계(sum), 승객 합계(count), 평균(mean)을 구한다.


```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic.head(1)
```

```python
titanic['agerange'] = pd.cut(titanic.Age, [0, 30, 60, 90], 
                             labels=['youth', 'middle-aged', 'old'])
```

```python
titanic.pivot_table(values='Survived', index='agerange', columns='Pclass', aggfunc='sum')
```

연령대와 승객등급별 생존자의 합계(sum)

```python
SurvSum = titanic.pivot_table(values='Survived', 
                    index='agerange', 
                    columns='Pclass', 
                    aggfunc='sum')
SurvSum
```

연령대와 승객등급별 생존자의 승객 합계(count)

```python
SurvCount = titanic.pivot_table(values='Survived', 
                    index='agerange', 
                    columns='Pclass', 
                    aggfunc='count')
SurvCount
```

```python
SurvMean = titanic.pivot_table(values='Survived', 
                    index='agerange', 
                    columns='Pclass', 
                    aggfunc='mean').applymap(lambda x: float('%.1f'%x))
SurvMean
```

```python
SurvResult = pd.concat([SurvSum, SurvCount, SurvMean],
                       axis=1, keys=['sum', 'count', 'mean'])
SurvResult
```
