```python
l1=['a','b','c','d','a','b','a','b']
l2=['b','a']
```

```python
from copy import copy
l3 = copy(l1)
for i in l3:
    for j in l2:
        if i == j:
            l1.remove(i)
l1
```

[실습] 사전형을 이용하여 알파벳의 빈도 구하기

```python
s = 'life is short, so python is easy.'
punct = ',. '
d = {}
```

```python
for c in s:
    if c in punct:
        continue
    d[c] = d.get(c, 0) + 1
```

```python
d
```

사전형을 이용하여 단어의 빈도 구하기

```python
s = 'life is short, so python is easy. python python easy'
punct = ',. '
d = {}
```

```python
for c in s.split(" "):
    c = c.strip(".")
    d[c] = d.get(c, 0) + 1
```

```python
for c in s.split(" "):
    c = c.strip(".")
    if c in d.keys():
        d[c] = d[c] + 1
    else:
        d[c] = 1
```

[실습] in.txt 파일을 읽들여서 열별로 각각 다른 파일로 저장하기

```python
data = []
with open('in.txt') as fin:
    for i, l in enumerate(fin.readlines()):
        if i in [0,1]:
            continue
        if i in [2]:
            names = l.strip().split(',')
            continue
        data.append(l.strip().split(','))
```

```python
outfiles = ['out1.txt', 'out2.txt', 'out3.txt']
columnset = zip(names, zip(*data))
for i, column in enumerate(columnset):
    with open(outfiles[i], 'w') as fo:
        fo.write(str(column))   
```
