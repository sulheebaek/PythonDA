## 수치 데이터의 정규화 또는 표준화
* 최소-최대 정규화<p>
$
X_{new} = {{X-min(X)} \over {max(X) - min(X)}}
$

* Z 점수 표준화<p>
$
X_{new} = {{X-Mean(X)} \over {StdDev(X)}}
$


```python
def zscore_standize(arr):
    return (arr - arr.mean())/arr.std()
```

```python
def minmax_normalize(arr):
    return (arr-arr.min())/(arr.max()-arr.min())
```

```python
concrete[:10]
```

```python
cnum = concrete.iloc[:8, :-1][:5]
cnum.plot()
```

```python
cnum.apply(minmax_normalize).plot()
```

```python
cnum.apply(zscore_standize).plot()
```

## 거리계산 방법의 적용
* 유클리디안 거리<p>
$
d(p, q) = \sqrt{ \left(p_1-q_1 \right) ^2 + \left( p_2-q_2 \right) ^2}
$
    

```python
import numpy as np
import operator
from numpy import linalg as la
from numpy import array, mat, sum, sqrt
```

```python
# 유클리디안 거리계산
def eDist(inA, inB):
    distance = np.sqrt(np.sum((inA-inB)**2,1))
    return distance
```

kNN 분류기

```python
def kNNClassifier(inX, trainSet, trainLabel, k=3, distMeasure=eDist):
    x = array(inX)
    tSet = array(trainSet)
    labels = array(trainLabel)
    dist = distMeasure(tSet, x)
    d = {}
    for l in labels[dist.argsort()][:k]:
        d[l] = d.get(l, 0)+1
    sortedlist = sorted(d.iteritems(), key=operator.itemgetter(1), reverse=True)
    return sortedlist[0][0]
```

```python
errorcount = 0.0
for i, dp in zip(mm.index, mm.values):
    answer = concrete['labels'][i]
    predicted = kNNClassifier(dp, mm, concrete['labels'], k=15)
    print('predicted result: %s' % predicted)
    print('train label: %s' % answer)
    if predicted != answer:
        errorcount += 1.0
print('errorcount = %.2f' % errorcount)
print('errorrate = %.2f' % (errorcount/len(mm.values)))
```
